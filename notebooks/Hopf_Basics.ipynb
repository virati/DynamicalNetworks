## Introduction
### Hopf dynamics
Hopf systems are pretty interesting because they exhibit a certain *change in qualitative behavior* that seems to be pretty wild. Imagine a pebble staying still while you heat it up. Higher and higher the temperature goes, you start to sweat, but the pebble is just sitting there.

Then, finally, once you reach a certain temperature the pebble just starts going in circles. Steady circles around where it was sitting just earlier, when the temperature was quite a bit less.

That change from being still to going around in circles is a crude analogy for what a Hopf bifurcation is: it's when something about your system's stability goes from being still to being cyclic or *oscillatory*.

## Basic Hopf bifurcation
Let's import what we need

```python
%reset
import numpy as np
import hopf_element as hel
import matplotlib.pyplot as plt
fig_size = [0,0]
fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
```

Once we've imported the libraries, with one custom one being the *hopf_element.py* library, we can plot its behavior at different values of its bifurcation parameter, or *critical variable* as we'll call it for this notebook.

```python
mu = 1
simpleNet = hel.HopfNet(center_freq=2,radius=0)
simpleNet.current_state = [-1.4,0]
simpleNet.plot_flow()
plt.show()
```

Here, we see the flow field of our Hopf system.

Let's see what would happen to this flow if we thought of those arrows as a lake and dropped a floatie somewhere off to the right. Where would that floatie end up.

```python
simpleNet.plot_flow(plot_traj=True)
plt.show()
```

If we zoom in a bit more, we can see that we eventually just collapse to the center at (0,0).

```python
simpleNet.plot_flow(plot_traj=True)
plt.subplot(2,1,1)
plt.xlim((-1,1))
plt.ylim((-1,1))
plt.show()
```

Now, we'll boost up our critical variable to see what happens to our floatie's trajectory.

```python
simpleNet.mu = 2
simpleNet.plot_flow(plot_traj=True)
plt.show()
```

And if we start from a different spot, like the bottom left, we see similar behaviors.

```python
simpleNet.plot_flow(plot_traj=True,state0=(-3,-3))
plt.show()
```

```python
simpleNet.mu = 2
simpleNet.plot_flow(plot_traj=True,state0=(0.01,0.01))
plt.show()
```

This strong drive for the system to settle onto oscillatory activity is called a *limit cycle*. In our case, this limit cycle is *stable* in that wherever we start the system wants to go to that limit cycle. With a *flip of a switch* we can actually make that *unstable*.
